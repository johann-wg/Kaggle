```python
plt.figure()
sns.stripplot(x='blue', y='ram', data=data, hue='price_range', dodge=True, jitter=True, palette='Set2')
plt.show()
```

```python
plt.figure(figsize=(15, 10))

for i in range(len(data.columns)):
    plt.subplot(3, 7, i+1)
    plt.boxplot(data[data.columns[i]])
    plt.title(data.columns[i])

plt.tight_layout()
plt.show()
```
