```python
import pandas as pd
import nltk
from nltk.tokenize import word_tokenize
nltk.download('punkt')

# 토큰화 함수 정의
def tokenize_sentences(sentence):
    return word_tokenize(sentence)

# 문장을 토큰화
train['tokenized'] = train['excerpt'].apply(tokenize_sentences)

# 고유 단어 사전 생성
word_set = set()
train['tokenized'].apply(lambda x: word_set.update(x))
word_to_index = {word: index for index, word in enumerate(word_set)}

# 텍스트를 숫자로 변환
train['encoded'] = train['tokenized'].apply(lambda x: [word_to_index[word] for word in x])

print(train)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# TF-IDF 변환기 초기화 및 변환
tfidf_vectorizer = TfidfVectorizer()
X_tfidf = tfidf_vectorizer.fit_transform(train['excerpt'])

# standard_error 값을 별도의 배열로 추출
X_standard_error = train['standard_error'].values.reshape(-1, 1)

# TF-IDF 벡터와 standard_error 값을 결합
X_combined = np.hstack((X_tfidf.toarray(), X_standard_error))
```
